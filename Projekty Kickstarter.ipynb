## Questions

2. Which project with a goal above $1,000 had the highest percentage of goal achievement (Declared / Goal)? How much money was declared?
3. Can you identify any trends in the success rate of projects over the years?

### 1. Which category has the highest success rate? How many projects were successful?
### Answer: Music, 24105 projects

```python
successful_categories = df[df['State'] == 'Successful'].groupby('Category')['ID'].count().to_frame().sort_values('ID', ascending = False).reset_index()
```

```python
successful_categories
```

```python
plt.figure(figsize=(10, 5))  
plt.barh(successful_categories['Category'], successful_categories['ID'], color = 'orange')
plt.title('Successful projects by categories')
plt.xlabel('Number of projects')
plt.xticks(rotation = 45)
plt.gca().invert_yaxis() 
plt.show()
```

### 2. Which project with a goal above 1 000 dollars had the highest percentage of goal achievement (Declared / Goal)? How much money was declared?
### Answer: Project Exploding Kittens, goal: 10 000 dollars, realisation: 87 824 %

```python
high_value = df[df['Goal'] > 1000]
```

```python
high_value['realisation_%'] = (df['Pledged'] / df['Goal']) * 100
```

```python
high_value['realisation_%'] = high_value['realisation_%'].round(2)
```

```python
high_value.sort_values('realisation_%', ascending = False).head(10)
```

### 3. Can you identify any trends in the success rate of projects over the years?
### Answer: The highest success rate was in years 2010 and 2011


```python
df['year'] = df['Deadline'].str.split(pat = '-').str[0]
```

```python
success = df[df['State'] == 'Successful']
```

```python
groupped = df.groupby('year').agg(total_projects=('ID', 'count'), successful_projects = ('State', lambda x: (x=='Successful').sum())).reset_index()
```

```python
groupped['%_succeed'] = (groupped['successful_projects'] / groupped['total_projects'] ) *100
```

```python
groupped
```

```python
plt.plot( groupped['year'], groupped['%_succeed'], color = 'green')
plt.show()
```
